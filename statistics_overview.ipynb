```python
#Standard Deviation

import math
def getsd(ls):
    summ=0
    x_mean=sum(ls)/len(ls)
    print("Mean :", x_mean)
    for l in ls:
        summ=summ+((l-x_mean)**2)
    print("Standard Deviation :",math.sqrt(summ/len(ls)))
```

```python
getsd([1,2,3,4,5])
```

```python
x=[3,4,5,6,4]
y=[2,3,4,5,1]
x1=[1,2,3,4,5]
y1=[1,2,3,4,5]
CR : 0.8320502943378437
def getcr(x, y):
    A=0
    B=0
    C=0
    x_mean=sum(x)/len(x)
    y_mean=sum(y)/len(y)
    for r in range(len(x)):#0 1 2 3 4
        A=A + (x[r]-x_mean)*(y[r]-y_mean)
        B=B + (x[r]-x_mean)**2
        C=C + (y[r]-y_mean)**2
    
    print("CR :", A/math.sqrt(B*C))
```

```python
getcr(x1,y1)
```

```python
import numpy as np
import statistics as sat
ar=np.array([2,3,4,5,6,7,6,6,2,2,2])
print(ar)

print("Mean :",np.mean(ar))
print("Median :",np.median(ar))
print("Max :",np.max(ar))
print("Min :",np.min(ar))
print("Standard Deviation: ", np.std(ar))
print("Mode  :", sat.mode(ar))
print("Range :", np.max(ar)-np.min(ar))
print(np.sort(ar))
```
